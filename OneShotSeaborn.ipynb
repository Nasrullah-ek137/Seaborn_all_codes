```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

data=sns.load_dataset('penguins').head(50)
#print(data)
```

```python
#LinePlot
sns.lineplot(x='bill_length_mm',y='bill_depth_mm',hue='sex',palette='Accent',data=data)
plt.title('This is LinePlot')
plt.grid()
plt.show()
```

```python
# Histogram plot
sns.displot(x='body_mass_g',bins=[2600,2800,3100,3400,3700,4000,4300,4600,4900,5000],color='m',kde=True,label='Histogram',data=data)
plt.title('This is Histogram')
plt.xlabel('prices')
plt.ylabel('counts')
plt.legend()
plt.show()
```

```python
# BarPlot
sns.barplot(x='island',y='body_mass_g',hue='sex',palette='husl',hue_order=['Female','Male'],order=['Dream','Biscoe','Torgersen'],data=data)
plt.title('This is Bar Plot')
plt.xlabel('Island land')
plt.ylabel('Price')
plt.show()
```

```python
# ScatterPlot
sns.scatterplot(x='bill_depth_mm',y='body_mass_g',hue='sex',sizes=(40,50),markers={'Female':'o','Male':'*'},data=data)
plt.title('This is Scatter Plot')
plt.xlabel('Quantity')
plt.ylabel('Price')
plt.show()
```

```python
data1=sns.load_dataset('anagrams').head(5)
del data1['attnr']
#print(data1)
posi=np.array([['A1','A2','A3','A4'],['B1','B2','B3','B4'],['C1','C2','C3','C4'],['D1','D2','D3','D4'],['E1','E2','E3','E4']])

sns.heatmap(data1,vmin=0,vmax=10,annot=posi,linewidth=4,linecolor='black',xticklabels=False,yticklabels=False,fmt='s')
plt.title('This is HeatMap')
plt.show()
```

```python
# CountPlot
data2=sns.load_dataset('tips')
#print(data2)
sns.countplot(x='sex',hue='smoker',palette='dark:g',data=data2)
plt.title('This is Count Plot')
plt.xlabel('Gender')
plt.ylabel('No.Of People')
plt.show()
```

```python
#VioLinPlot
sns.violinplot(x='day',y='total_bill',hue='sex',data=data2)
plt.title('This is Violin Plot')
plt.xlabel('No.Of Days')
plt.ylabel('Total Bill')
plt.show()
```

```python
# PairPlot
sns.pairplot(data2,hue='sex',markers=['*','o'])
plt.show()
```

```python
#Stripplot
sns.stripplot(x='day',y='total_bill',hue='sex',jitter=5,data=data2)
plt.title('This is Strip Plot')
plt.xlabel('No.Of Days')
plt.ylabel('Total Bill')
plt.show()
```

```python
# BoxPlot
sns.boxplot(x='day',y='total_bill',hue='sex',showmeans=True,data=data2)
plt.title('This is Box Plot')
plt.xlabel('No.Of Days')
plt.ylabel('Total Bill')
plt.show()
```

```python
#CatPlot
sns.catplot(x='size',y='tip',kind='bar',hue='sex',palette='Accent',data=data2)
plt.title('This is Cat Plot')
plt.xlabel('Sizee')
plt.ylabel('Tips')
plt.show()
```

```python
#MultiplePlot
mp=sns.FacetGrid(data2,col='sex',hue='day')
mp.map(plt.scatter,'total_bill','tip').add_legend()
plt.show()
```
